```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr
import seaborn as sns

data = pd.read_csv('refined_data.csv')
dataDF = data[['price','address','city','zipcode','beds','baths','floor_size','Lot','Property_Type','year_built','Description']]
```

# Question 1:

Properties from which city in the bay area, on average, has the best price to features ratio when  the features in consideration are number of bedrooms, number of bathrooms, floor size, and lot size?

1) Cleaning the data

```python
# secondary cleaning of the data

clean_dataDF = dataDF.loc[(dataDF['city'] != 'None') 
                          & (dataDF['beds'] != 'None') 
                          & (dataDF['baths'] != 'None') 
                          & (dataDF['floor_size'] != 'None') 
                          & (dataDF['Lot'] != 'None')]

Lot_size = []
Floor_size = []

for each in clean_dataDF['Lot'].values.tolist():
    spliteach = each.split(' ')
    splitnum = spliteach[0].split(',')
    num = ""
    for each in splitnum:
        num += each
    
    if spliteach[1] == 'sqft' or float(num) > 100:
        Lot_size.append(float(num))
    elif spliteach[1] == 'acres' or spliteach[1] == 'acre':
        Lot_size.append(float(num) * 43560)
        
for each in clean_dataDF['floor_size'].values.tolist():
    splitnum = each.split(',')
    num = ""
    for each in splitnum:
        num += each
    Floor_size.append(float(num))

# changing all quantitative values to float type
clean_dataDF['baths'] = clean_dataDF['baths'].astype(float)
clean_dataDF['beds'] = clean_dataDF['beds'].astype(float)

clean_dataDF['Lot'] = Lot_size
clean_dataDF['floor_size'] = Floor_size
clean_dataDF.reset_index()

# remove all properties in cities with 10 or less rows
clean_dataDF_count = clean_dataDF.groupby(['city'])[['price','floor_size','beds','baths','Lot']].count()
clean_dataDF_count = clean_dataDF_count.loc[clean_dataDF_count['price'] > 10]
clean_dataDF = clean_dataDF.merge(clean_dataDF_count, how='right', on='city')
clean_dataDF = clean_dataDF[['city','address','price_x','beds_x','baths_x','floor_size_x','Lot_x']]
clean_dataDF.columns=['city','address','price','beds','baths','floor_size_sqft','lot_size_sqft']

# viewing the data
#clean_dataDF
```

2) Analyzing the data

My method for answering question 1 is as follow:

- A dataframe is first created showing the average property price, number of bedrooms, number of bathrooms, floor size, and lot size for each city. (shown below)

```python
clean_dataDF_mean = clean_dataDF.groupby(['city'])[['price','beds','baths','floor_size_sqft','lot_size_sqft']].mean().round(2)
clean_dataDF_mean
```

- Using the data above, a new dataframe is created with four unique columns ('beds', 'baths', 'floor_size', 'lot_size'), which would be calculated using the four formulas below, respectively:

      beds: (average property price of each city)/(average number of bedrooms for properties in each city) 
    
      baths: (average property price of each city)/(average number of bathrooms for properties in each city) 
    
      floor_size: (average property price of each city)/(average floor size for properties in each city) 
    
      lot_size: (average property price of each city)/(average lot size for properties in each city) 
    
  The purpose of this step is to consider how much it cost, on average, for each of the features - bedrooms, bathrooms, floor size, lot size -, when each feature is considered individually, for each city. Basically, it is a ratio of average total cost to average feature.

```python
#index dataframe where the calculated indexes for each variable of each city will be displayed
ratioDF = pd.DataFrame(columns=['city','price:beds','price:baths','price:floor_size','price:lot_size'])
ratioDF['city'] = clean_dataDF_mean.index

for col in range(len(clean_dataDF_mean.columns)-1):
    for row in range(len(clean_dataDF_mean.iloc[:,col+1])):
        ratioDF.iloc[row,col+1] = clean_dataDF_mean.iloc[row,0]/clean_dataDF_mean.iloc[row,col+1]   

ratioDF
```

- In each column, the values are normalized. Since the values in each column are suppose to show the ratio of average property price to average feature count, normalizing the values in each feature column would show how high or how low this particular ratio is for a city when compared to other cities for the feature. Basically, the closer the normalized value is to 1 for a feature at a city, the higher the ratio of average property price to average feature count, or, the more you are paying, for the features at that city.   

```python
ratioDF_norm = ratioDF.copy()
ratioDF_norm.columns = ['city','beds_norm','baths_norm','floor_size_norm','lot_size_norm']

for col in range(len(ratioDF_norm.columns) - 1):
    var_min = ratioDF_norm.iloc[:,col+1].min()
    var_max = ratioDF_norm.iloc[:,col+1].max()
    
    for row in range(len(ratioDF_norm.iloc[:,col+1])):
        ratioDF_norm.iloc[row,col+1] = (ratioDF_norm.iloc[row,col+1]-var_min)/(var_max-var_min)

ratioDF_norm
```

- At the end, the normalized ratios for each city are averaged out and placed into an average_norm_score column. The lower the average score for a city, the lower the price you are paying for each feature, on average, for properties in that city.

```python
ratioDF_norm['avg_norm_score'] = (ratioDF_norm['beds_norm'] 
+ ratioDF_norm['baths_norm'] + ratioDF_norm['floor_size_norm']
+ ratioDF_norm['lot_size_norm']) / 4

ratioDF_norm
```

3) Visualization

Remember, the higher the index score (out of 4) for a city, the more, on average, you are gaining from the money you pay for a property in that particular city.

```python
fig, axs = plt.subplots(5,1,figsize=(20,30))

axs[0].bar(ratioDF_norm['city'],ratioDF_norm['beds_norm'], color='red')
axs[0].set_ylabel('Normalized Ratio')
axs[0].set_xlabel('Cities')
axs[0].set_title('Normalized Ratios of Average Property Price to Average Number of Bedrooms By Cities')

axs[1].bar(ratioDF_norm['city'],ratioDF_norm['baths_norm'], color='yellow')
axs[1].set_ylabel('Normalized Ratio')
axs[1].set_xlabel('Cities')
axs[1].set_title('Normalized Ratios of Average Property Price to Average Number of Bathrooms By Cities')

axs[2].bar(ratioDF_norm['city'],ratioDF_norm['floor_size_norm'], color='green')
axs[2].set_ylabel('Normalized Ratio')
axs[2].set_xlabel('Cities')
axs[2].set_title('Normalized Ratios of Average Property Price to Average Floor Size By Cities')

axs[3].bar(ratioDF_norm['city'],ratioDF_norm['lot_size_norm'], color='blue')
axs[3].set_ylabel('Normalized Ratio')
axs[3].set_xlabel('Cities')
axs[3].set_title('Normalized Ratios of Average Property Price to Average Lot Size By Cities')

axs[4].bar(ratioDF_norm['city'],ratioDF_norm['avg_norm_score'], color='black')
axs[4].set_ylabel('Averaged Normalized Ratio')
axs[4].set_xlabel('Cities')
axs[4].set_title('Average Across all Normalized Ratios for All Features By Cities')

plt.show()
```

# Result

As it is very clearly shown, at a mean normalized ratio of 0, the properties in Vacaville, has the lowest average property price to average features ratios across all of the four features (average number of bedrooms, number of bathrooms, floor size, and lot size). This means that Vacaville properties, on average, has the best price deals for the aforementioned features. On the other hand, properties in San Francisco, on average, would make you pay the most for the aforementioned features. It is also noteworthy that Palo Alto (average norm ratio of 0.85031) and San Francisco (average norm ratio of 0.934452) are not only ranked as two of the highest overall ratios, but they are also much higher than even the third highest ratio, Sunnyvale (average norm ratio of 0.32917). This means that the average property features in Palo Alto and San Francisco are a lot more expensive compared to any of the other cities in the bay area.

# Question 2:

How much higher or lower would the average property prices of each city in the bay area be predicted to be if they were to be estimated using only what we know about the bedroom, bathroom, floor size, and lot size features, as well as the price of the properties in the most expensive bay area city?

It is not possible to use a single pricing scheme for bedrooms, bathrooms, floor size, and lot size because different realestate agents would price the features of a property differently. Hence, for this question, my method will be to use a linear regression model, using scikit, that will use the four aforementioned features from the most expensive city for fitting. For this case, it is clear that San Francisco is the most expensive city in the bay area, as shown in question 1. Hence, the feature data -using number of bedrooms, number of bathrooms, floor size, lot size- from San Francisco will be used to fit a multiple linear regression model. 

1) Analyzing the Data

My methods for analyzing the data will be as follow:

- Using the sklearn.linear_model package, a multiple linear regression model is created using only the San Francisco property data. The model will be used to predict property prices using the number of bedrooms, number of bathrooms, property floor size, and property lot size as predictor variables.

```python
# linear regression model
lrdata = clean_dataDF.loc[(clean_dataDF['city'] == '\xa0San Francisco')]
lrdata_y = lrdata['price']
lrdata_x = lrdata[['beds','baths','floor_size_sqft','lot_size_sqft']]
lrdata_y_array = np.array(lrdata_y)
lrdata_x_array = np.array(lrdata_x)

lrmodel = lr().fit(lrdata_x_array,lrdata_y_array)

# model's coefficient of determination
print('coefficient of determination: ', lrmodel.score(lrdata_x_array,lrdata_y_array))
print('intercept: ', lrmodel.intercept_)
print('slopes: ', lrmodel.coef_)
```

 - predicting the property prices for every other city in the dataset using the linear regression model that was fitted to the San Francisco data.

```python
city_names = clean_dataDF.groupby(['city']).count().index

pred_price_DF_all = pd.DataFrame(columns=['city','address','actual price','predicted price','error'])

for each in city_names:
    df = clean_dataDF.loc[clean_dataDF['city'] == each]
    df_x = np.array(df[['beds','baths','floor_size_sqft','lot_size_sqft']])
    price_pred = pd.Series(lrmodel.predict(df_x))
    pred_price_DF = pd.concat([df[['city','address','price']].reset_index(), price_pred], axis=1)
    pred_price_DF = pred_price_DF.iloc[:,1:5]
    pred_price_DF.columns = ['city','address','actual price','predicted price']
    pred_price_DF['error'] = pred_price_DF['predicted price'] - pred_price_DF['actual price']
    pred_price_DF_all = pd.concat([pred_price_DF_all,pred_price_DF])

pred_price_DF_all.round(1)
```

- averaging the data

```python
pred_price_DF_average = pred_price_DF_all.groupby(['city'])[['actual price','predicted price','error']].mean().round(2)
pred_price_DF_average
```

 3) Visualizing the Data

```python
pred_price_DF_average.plot(kind='bar', figsize=(15,7))
plt.legend(loc='upper center')
plt.xlabel('Cities')
plt.ylabel('Price')
plt.title('Linear Regression Results - Actual Vs Predicted Property Prices by City')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

# Result

As shown by the graph and the charts, using the property data from only San Francisco to predict the average property price of other bay area cities resulted in all other cities having a significantly higher predicted average property price than their actual average prices. Notably, Daly City, Fremont, and Vacaville have the most significant error in prediction by a large margin.

# Question 4
Visualize the percentage of different property types for each city. Which cities have the most of which type of properties for sale?

1) clean data

```python
clean_dataDF_proptype = dataDF.loc[dataDF['Property_Type'] != 'None']
```

```python
proptype_types = clean_dataDF_proptype.groupby(['Property_Type']).count()
proptype_indexes = proptype_types.index
col = ['City']

for each in proptype_indexes:
    col.append(each)

proptype_DF = pd.DataFrame(columns=col)
```

- Fill in the City column of the dataframe with the appropriate city names that have at least more than 10 rows

```python
proptype_cities = clean_dataDF_proptype.groupby(['city']).count()
proptype_cities = proptype_cities.loc[proptype_cities['Property_Type'] > 10]
pindex = proptype_cities.index
```

- calculate the count of each property type for each city and fill in any missing columns with 0

```python
for city in pindex:
    row = [[]]
    citiesdata = clean_dataDF_proptype.loc[clean_dataDF_proptype['city'] == city]
    proptype_count = citiesdata.groupby(['Property_Type'])['city'].count()
    
    colnames = ['City']
    for each in proptype_count.index:
        colnames.append(each)
    
    row[0].append(city)
    for each in proptype_count:
        row[0].append(each)
       
    df_row = pd.DataFrame(row, columns=colnames)
    proptype_DF = pd.concat([proptype_DF, df_row], ignore_index=True, sort=False)

proptype_DF = proptype_DF.fillna(0)
proptype_DF
```

- convert all numbers to percentages

```python
proptype_DF_perc = proptype_DF.copy()
proptype_DF_perc.iloc[:,1:] = proptype_DF_perc.iloc[:,1:].astype(float) 

for col in range(len(proptype_DF_perc.columns)-1):
    proptype_DF_perc.iloc[:,col+1] = proptype_DF.iloc[:,col+1] / proptype_DF.iloc[:,1:].sum(axis=1)

proptype_DF_perc
```

3) Visualize the data

```python
# Property types for all cities put into lists
cities = []
labels = []
sizes = []

for row in range(len(proptype_DF_perc)):
    rowlabels = []
    rowsizes = []
    
    cities.append(proptype_DF_perc.iloc[row,0][1:])
    
    for col in range(len(proptype_DF_perc.columns)-1):
        if proptype_DF_perc.iloc[row,col+1] > 0:
            rowlabels.append(proptype_DF_perc.columns[col+1])
            rowsizes.append(proptype_DF_perc.iloc[row,col+1])    
    
    labels.append(rowlabels)
    sizes.append(rowsizes) 
```
