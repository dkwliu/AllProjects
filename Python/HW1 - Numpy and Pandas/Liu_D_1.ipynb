## Problem 1 Python Basics (30 pts)

Your will first write a function PasswordCheck() that determines whether or not a password is good. We will define a good password to be a one that is at least 8 characters long and contains at least one uppercase letter, at least one lowercase letter, and at least one number. Your function should return True if the password passed to it as its only parameter is good. Otherwise it should return False. (10pts)

Then write another function called randomePassword(). The random password shoudl have a random length of between 7 to 11 characters. Each character should be randomly selected from position 33 to 126 in the ASCII table.Your function will not take any parameters. It will return the randomly generated password as its only result. (10pts)

A random character can be generated by:

    randomChar = chr(randint(MIN_ASCII, MAX_ASCII)),
    
where ASCII is the traditional name for encoding system. MIN_ASCII = 33, MAX_ASCII = 126. You also need to import **random** moudle. 

In your main function, use randomePassword() to generate a password and display it. Then us PasswordCheck() to determine if this is a good password. Count and display the number of attempts that were needed until a good password was generated.  Ensure that your main program only runs when your solution has not been imported into another file. (10pts)


```python
import random as rand

def PasswordCheck(pw):
    length = False
    cap = False
    low = False
    num = False
    
    if len(pw) >= 8:
        length = True
    
    for each in pw:
        if each.isupper():
            cap = True
        elif each.islower():
            low = True
        elif each.isnumeric():
            num = True
    
    if length == True & cap == True & low == True & num == True:
        return True
    else:
        return False

def randomePassword():
    length = rand.randint(7,11)
    pw = ""
    
    for c in range(length):
        asciiInt = rand.randint(33,126)
        char = chr(asciiInt)
        
        pw = pw + char
    
    return pw

def main():
    count = 0
    goodPw = False
    pw = []
    
    while (goodPw == False):
        randomPw = randomePassword()
        
        pw.insert(count, randomPw + "\n")
        
        if PasswordCheck(randomPw) == True:
            goodPw = True
            
        count += 1
        
        #Main method only runs when solution isn't yet imported to another file.
        #
        #Main method will stop running as soon as a good password is
        #detected and the solutions are imported to a txt file.
        if goodPw == True:
            pwFile = open('pwFile.txt','w')
            pwFile.writelines(pw)
            pwFile.write(str(count) + " attempt count.")
            pwFile.close()
        
        print(randomPw, " --- ",goodPw)  
    
    print(count,"attempt count.")
    
main()
```

## Problem 2 Numpy Basics (15 pts)
Create two random **uniform** matrices of same shape (5,3), named A and B. 

Check if they are equal. (5pts)


```python
import numpy as np

A = np.random.randn(5,3)
B = np.random.randn(5,3)
print(A,"\n")
print(B,"\n")

print("Are they equal?:", np.array_equal(A,B))
```

In each row of matrix A, please replace the maximum value by 0. Then print these two updated matrices. (10pts)

```python
def replaceMax(array):
    for each in array:
        maxVal = max(each)
        for val in range(len(each)):
            each[val] = np.where(each[val] == maxVal, 0, each[val])
        
replaceMax(A)
replaceMax(B)

print("Modified Array A:\n", A)
print("Modified Array B:\n", B)
```

## Porblem 4 Numpy Random Walk (15 pts)

Consider a simple random walk strating from 0 with steps of 1, 0 and -1 occuring with probablity 0.5, 0.3 and 0.2 respectively. Then we implement a single random walk with 100 steps using the built-in random module in Numpy. Then plot out the cumulative steps using matplotlib.pyplot module. 

```python
import matplotlib.pyplot as plt

steps = np.random.choice([1,0,-1], size=100, p=[0.5,0.3,0.2])
allsteps = steps.cumsum()

plt.plot(allsteps)
```
