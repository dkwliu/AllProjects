# Problem 5: Simple Linear Regression (20 pts)

Given the following data:

x = [30., 40., 40., 50., 60., 70., 70., 70., 80., 90.] List of advertising investment in $1000

y = [184.4, 279.1, 244.0, 314.2, 382.2, 450.2, 423.6, 410.2, 500.4, 505.3] List of corresponding sales in $1000

The regression function is defined as h(x) = t0 + t1* x. 

Please estimate the value of t0 and t1 using **Least Square method**. Print the result of t0 and t1 and R^2 (R square).  (20 pts)

**Hint:**
The fomulas can be found in the following:
    ![title](LeastSquare.png)

```python
y = [184.4, 279.1, 244.0, 314.2, 382.2, 450.2, 423.6, 410.2, 500.4, 505.3]
x = [30., 40., 40., 50., 60., 70., 70., 70., 80., 90.]

y_mean = np.mean(y)
x_mean = np.mean(x)

y_error = [(each - y_mean) for each in y] 
x_error = [(each - x_mean) for each in x]

y_error_sq = [(each * each) for each in y_error]
x_error_sq = [(each * each) for each in x_error]


xy_error = np.ndarray.tolist(np.array(x_error) * np.array(y_error))

r = np.sum(xy_error)/np.sqrt(np.sum(x_error_sq) * np.sum(y_error_sq))

SDx = np.sqrt(np.sum(x_error_sq)/len(x))
SDy = np.sqrt(np.sum(y_error_sq)/len(y))

b1 = r * SDy/SDx

b0 = y_mean - (b1 * x_mean)

print("b1 is approximately", b1,", b0 is approximately", b0,", and r^2 is approximately", r**2)
```
