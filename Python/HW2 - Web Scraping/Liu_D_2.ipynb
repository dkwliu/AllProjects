# Problem 1 Scraping contents in HTML using Beautiful Soup (50 pts)

Given the first page of search result of apartment/condos in Fremont, CA,

    https://www.zillow.com/fremont-ca/condos/?searchQueryState={%22pagination%22:{},%22usersSearchTerm%22:%22Fremont,%20CA%22,%22mapBounds%22:{%22west%22:-122.20828434277348,%22east%22:-121.8086566572266,%22south%22:37.31074450283708,%22north%22:37.747463364882506},%22regionSelection%22:[{%22regionId%22:11540,%22regionType%22:6}],%22isMapVisible%22:true,%22mapZoom%22:11,%22filterState%22:{%22isSingleFamily%22:{%22value%22:false},%22isManufactured%22:{%22value%22:false},%22isMultiFamily%22:{%22value%22:false},%22isLotLand%22:{%22value%22:false},%22isTownhouse%22:{%22value%22:false}},%22isListVisible%22:true}
    

(in the first page of search results) please scrape info of the first 20 listing homes: **listing price, short description below the price**. For example, the first search result has the description as "Residence 2 Plan, Apex at Mission Stevenson".  Store all the information in a DataFrame. Note, this DataFrame contains 2 columns.  **Notice:** please use Beautiful Soup lib. 

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd
import ssl

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

zillowHomes  = pd.DataFrame(columns=['Address_Details','Price'])

url = "https://www.zillow.com/fremont-ca/condos/?searchQueryState={%22pagination%22:{},%22usersSearchTerm%22:%22Fremont,%20CA%22,%22mapBounds%22:{%22west%22:-122.20828434277348,%22east%22:-121.8086566572266,%22south%22:37.31074450283708,%22north%22:37.747463364882506},%22regionSelection%22:[{%22regionId%22:11540,%22regionType%22:6}],%22isMapVisible%22:true,%22mapZoom%22:11,%22filterState%22:{%22isSingleFamily%22:{%22value%22:false},%22isManufactured%22:{%22value%22:false},%22isMultiFamily%22:{%22value%22:false},%22isLotLand%22:{%22value%22:false},%22isTownhouse%22:{%22value%22:false}},%22isListVisible%22:true}"
    
webReq = Request(url, headers={'User-Agent':'Mozilla/5.0'})
content = urlopen(webReq)
soup = bs(content, 'html.parser')

addrDet = soup.findAll('address',attrs={'class':'list-card-addr'})
price = soup.findAll('div',attrs={'class':'list-card-price'})

for each in range(20):
    zillowHomes.loc[each,'Address_Details'] = addrDet[each].get_text()
    zillowHomes.loc[each,'Price'] = price[each].get_text()

zillowHomes

```

# Problem 2 Scraping contents in HTML using Regax  (20 pts)

Given this website: 
https://www.zillow.com/homedetails/5651-Spry-Cmn-Fremont-CA-94538/25045996_zpid/  , 

please scrape the **HOA and Price/sqft**.

**Notice:** please use RE lib. 

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs
import ssl
import re

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = "https://www.zillow.com/homedetails/5651-Spry-Cmn-Fremont-CA-94538/25045996_zpid/"
req = Request(url, headers={'User-Agent':'Mozilla/5.0'})
content = urlopen(req)
soup = bs(content, "html.parser")

hoa = re.findall('<span class="ds-standard-label ds-home-fact-label">HOA:</span><span class="ds-body ds-home-fact-value">(.*?/month)</span>', str(soup))
pricesqft = re.findall('<span class="ds-standard-label ds-home-fact-label">Price/sqft:</span><span class="ds-body ds-home-fact-value">(.[0-9]+)</span>', str(soup))

print('HOA:', hoa[0])
print('Price/sqft:', pricesqft[0])
```

# Problem 3 json API  (30 pts)

This website shows the data science related jobs in json file. 

https://jobs.github.com/positions.json?description=data+scientists&location=


(1) Please use _request_ lib and make a _GET_ request to scrape the data in json, and transfer them into a dictionary (more specifically, it is a list of dictionaries). (10 pts)

```python
import requests
import json

url = "https://jobs.github.com/positions.json?description=data+scientists&location="

gitData = requests.get(url)

gitDic = gitData.json()

print(gitDic) # The list of dictionary
```

(2) Can you make a DataFrame out of this dictionary, where the columns include: 'type','url','company','location','title','description'. Please name this DataFrame jobs. (10 pts)

```python
import pandas

jobs = pd.DataFrame(gitDic,columns=['type','url','company','location','title','description'])
jobs
```

(3) **Challenge Question**: filter out the jobs in California (any city in California). (**Hint**: Please use RE lib.)(5 pts)

```python
#filtered for the jobs in California

filter = re.findall('([a-zA-Z].*[a-zA-Z]*, CA)', str(jobs['location']))

print('The following rows are filtered for cities in CA')
jobs.iloc[:,:][[True if i in filter else False for i in jobs['location']]]
```
